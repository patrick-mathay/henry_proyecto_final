```python
# Initialize an empty list to store parsed JSON objects
data_list = []

# Open the JSON file and read it line by line with UTF-8 encoding
with open(r'C:\Users\Owner\Desktop\proyecto_final\data_yelp\review.json', 'r', encoding='utf-8') as json_file:
    for line in json_file:
        # Parse each line as JSON and append it to the list
        data_list.append(json.loads(line))

# Convert the list of JSON objects into a DataFrame
df_review = pd.DataFrame(data_list)
```
