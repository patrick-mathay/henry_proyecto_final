```python
import pandas as pd
```

```python
dfy = pd.read_csv(r'dataset_business_final_yelp.csv')
dfm = pd.read_csv(r'dataset_final_business_maps.csv')
```

Normalización de dfm Maps Business

```python
dfm['latitude'] = dfm['latitude'].round(6)
dfm['longitude'] = dfm['longitude'].round(6)
```

```python
# Crear la nueva columna 'attributes' combinando las otras tres columnas
dfm['attributesM'] = dfm[['Accessibility', 'Planning', 'Service options']].apply(lambda x: ','.join(x.dropna().astype(str)), axis=1)
# Eliminar 'SinDato' de los strings en la columna 'attributes'
dfm['attributesM'] = dfm['attributesM'].str.replace('SinDato', '').str.replace(',,', ',').str.strip(',')
```

```python
# Eliminar las columnas originales
dfm = dfm.drop(columns=['Accessibility', 'Planning', 'Service options'])
```

```python
# Eliminar otras columnas originales
dfm = dfm.drop(columns=['url', 'relative_results', 'state'])
```

```python
# cambiar el nombre a la columna 'avg_rating' para especificar que es de google
dfm = dfm.rename(columns={'avg_rating': 'avg_ratingM'})
dfm = dfm.rename(columns={'num_of_reviews': 'num_of_reviewsM'})
```

```python
# Especifica el nuevo orden de las columnas
nuevo_orden = ['gmap_id', 'name', 'category_id', 'business_category', 'latitude', 'longitude', 'adress', 'avg_ratingM', 'num_of_reviewsM', 'hours', 'attributesM' ]

# Reordena las columnas
dfm = dfm.reindex(columns=nuevo_orden)
```

Normalización de datos de Yelp Business

```python
dfy['latitude'] = dfy['latitude'].round(6)
dfy['longitude'] = dfy['longitude'].round(6)
```

```python
valores_unicos = dfy['stars'].unique()
print(valores_unicos)
```

```python
# Eliminar las columnas originales
dfy = dfy.drop(columns=['city', 'ubicacion', 'postal_code'])
```

```python
# Renombrar la columna 'business_id'
dfy = dfy.rename(columns={'business_id': 'yelp_id'})
```

```python
dfy = dfy.rename(columns={'review_count': 'num_of_reviews', 'attributes': 'attributes'})
```

```python
# Especifica el nuevo orden de las columnas
nuevo_orden = ['yelp_id', 'name', 'category_id', 'business_category', 'latitude', 'longitude', 'adress', 'stars', 'num_of_reviews', 'hours', 'attributes' ]

# Reordena las columnas
dfy = dfy.reindex(columns=nuevo_orden)
```

Se unifican los archivos para crear la clave business_id y la tabla de Business

```python
# Crea la columna 'business_id' uniendo latitud y longitud
dfm['business_id'] = dfm[['longitude', 'latitude']].apply(lambda row: f"{row['longitude']}-{row['latitude']}", axis=1)
```

```python
# Crea la columna 'business_id' uniendo latitud y longitud
dfy['business_id'] = dfy[['longitude', 'latitude']].apply(lambda row: f"{row['longitude']}-{row['latitude']}", axis=1)
```

```python
# Comprueba el tipo de datos de las columnas business_id
print("Tipo de dato de dfm['business_id']:", dfm['business_id'].dtype)
print("Tipo de dato de dfy['business_id']:", dfy['business_id'].dtype)
```

Creo la tabla Business

```python
df_concatenado = pd.concat([dfm, dfy])
```

```python
nuevo_orden = ['business_id', 'latitude', 'longitude', 'name', 'category_id', 'business_category', 'adress', 'gmap_id', 'yelp_id','avg_rating', 'stars', 'num_of_reviews', 'num_of_reviews', 'attributesM', 'attributesY', 'hours' ]
# Reordena las columnas
df_concatenado = df_concatenado.reindex(columns=nuevo_orden)
```

```python
# Traemos sólo las columnas 'business_id', 'latitude', 'longitude'y 'name' que formarán la tabla nueva de Business
df_business_ids = df_concatenado[['business_id', 'latitude', 'longitude', 'name']]
```
