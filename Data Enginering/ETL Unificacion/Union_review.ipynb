```python
import pandas as pd
```

Normalización de datos de Yelp Reviews

```python
dfy = pd.read_csv(r'dataset_review_final_yelp.csv')
```

```python
dfy
```

```python
# Contar cuántos valores únicos hay en la columna 'stars'
unique_y = dfy['stars'].nunique()
# cambiar el tipo de dato a tipo entero
dfy['stars'] = dfy['stars'].astype(int)
```

```python
# Crear la columna 'rating' y caragar con la suma de las otras columnas
dfy['rating'] = dfy['useful'] + dfy['funny'] + dfy['cool'] + dfy['stars']
```

```python
# eliminar las columnas
columnas_a_eliminar = ['useful', 'funny', 'cool', 'stars']
dfy = dfy.drop(columnas_a_eliminar, axis=1)
```

```python
# Cambiar el nombre de la columna business_id por yelp_id
dfy = dfy.rename(columns={'business_id': 'yelp_id'})
```

```python
# Crear la columna 'platform' y poblarla con la letra Y
dfy['platform'] = 'Y'
```

```python
dfy
```

Normalización de datos de Maps Review

```python
dfm = pd.read_csv(r'dataset_review_final_maps.csv')
```

```python
dfm
```

```python
# Contar cuántos 'SinDato' hay en cada columna
count_time_resp = (dfm['time_resp'] == 'SinDato').sum()
count_text_resp = (dfm['text_resp'] == 'SinDato').sum()
print(f"Cantidad de 'SinDato' en 'time_resp': {count_time_resp}")
print(f"Cantidad de 'SinDato' en 'text_resp': {count_text_resp}")
```

```python
# eliminar las columnas
columnas_a_eliminar = ['time_resp', 'text_resp', 'name']
dfm = dfm.drop(columnas_a_eliminar, axis=1)
```

```python
# Cambiar el nombre de la columna time por date
dfm = dfm.rename(columns={'time': 'date'})
```

```python
# Crear la columna 'review_id' y poblarla con una clave única
dfm['review_id'] = range(1, len(dfm) + 1)
```

```python
# Crear la columna 'platform' y poblarla con la letra M
dfm['platform'] = 'M'
```

```python
# Reordenar las columnas
nuevo_orden = ['review_id', 'user_id', 'gmap_id', 'text', 'date', 'rating', 'platform']  
dfm = dfm[nuevo_orden]
```

Se unifican los datasets de review Yelp y Maps

Para eso, necesitamos importar los datasets de Business tanto de Yelp como de Maps

```python
df_BY = pd.read_csv(r'Business_Yelp.csv')
```

```python
df_BM = pd.read_csv(r'Business_Maps.csv')
```

Renombramos los datasets de Reviews

```python
# Dataframe Reviews Yelp
df_RY = dfy
# Dataframe Reviews Maps
df_RM = dfm
```

```python
#df_RM = pd.read_csv(r'C:\Users\Sofi\Desktop\Henry\LABS\Borrador_PF\PF\Data\Union_datasets\Review\review_maps_normalized.csv')
```

Agregar columna 'business_id' al dataset de 'review_yelp_normalized'

```python
df_Ryelp_con_business_id = df_RY.merge(df_BY[['yelp_id', 'business_id']], on='yelp_id', how='left')
```

```python
df_Ryelp_con_business_id
```

Agregar columna 'business_id' al dataset de 'review_maps_normalized'

```python
df_RM_con_business_id = df_RM.merge(df_BM[['gmap_id', 'business_id']], on='gmap_id', how='left')
```

```python
# Sumar los nulos que equivalen a los registros que no tuvieron coincidencias.
df_RM_con_business_id.isnull().sum().sum()
```

```python
# Eliminamos los nulos ya que son registros que no se corresponden con restaurantes.
df_sin_nan_business_id = df_RM_con_business_id.dropna(subset=['business_id'])
```

```python
df_sin_nan_business_id
```

Limpiar y ordenar datasets antes de unificar

```python
# Cambiar el nombre de la columna yelp_id por platform_id
df_reviews_yelp = df_Ryelp_con_business_id.rename(columns={'yelp_id': 'platform_id'})
```

```python
# Cambiar el nombre de la columna gmap_id por platform_id
df_reviews_maps = df_sin_nan_business_id.rename(columns={'gmap_id': 'platform_id'})
```

Unificar datasets

```python
df_reviews_union = pd.concat([df_reviews_maps, df_reviews_yelp], ignore_index=True)

# Verifica las primeras filas del DataFrame resultante
df_reviews_union
```

```python
# Reordenar las columnas
nuevo_orden = ['review_id', 'user_id', 'platform_id', 'platform', 'business_id', 'text', 'date', 'rating']  

# Reordena las columnas
df_reviews_union = df_reviews_union[nuevo_orden]
```

```python
# Exportar dataset
df_reviews_union.to_csv('reviews_union.csv', index=False)
```
