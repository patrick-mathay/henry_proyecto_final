Crear un funcion para analizar el tipo de datos proporcionados

```python
# Funcion para revisar el tipo de dato de cada variable del dataset 
def tipo_datos(df):
    mi_dict = {"nombre_campo": [], "tipo_datos": []}
    for columna in df.columns:
        mi_dict["nombre_campo"].append(columna)
        mi_dict["tipo_datos"].append(df[columna].apply(type).unique())
    df_info = pd.DataFrame(mi_dict)
    return df_info
```

### Variable pics

### Variable text

* La variable muestra el texto de la revisión, en los valores nulos muestra revisiones que no adjuntaron texto, pero si calificaron su rating
* Se completa valores nulos conn el valor "SinDato"

```python
# Se Completa los valores nulos con el valor "SinDato"
df_florida["text"].fillna("SinDato", inplace=True)
df_florida
```

### Variable resp

* La variable resp muestra la respuesta del negocio a la reseña, hay un diccionario que muestra el tiempo (time) y respuesta (text)
* Los datos se encuentran anidados, por lo tanto se procede a desanidar en 02 columnas nuevas, time_resp y text_resp

```python
# Se normaliza la columna "resp" del DataFrame df_florida en filas separadas
df_resp_normalized = pd.json_normalize(df_florida['resp'])
df_resp_normalized
```

```python
# Se Completa los valores nulos con el valor "SinDato"
df_resp_normalized["text"].fillna("SinDato", inplace=True)
df_resp_normalized
```

```python
# Se Completa los valores nulos con el valor "SinDato"
df_resp_normalized["time"].fillna(0, inplace=True)
df_resp_normalized
```

```python
# se crea una funcion para cambiar la visualizacion de la columna time de cientifica a numero.
def format_time(time):
    return '%.0f' % time

# aplicamos la funcion a la columna 
df_resp_normalized['time'] = df_resp_normalized['time'].apply(format_time)

# se cambia el nombre de columnas
df_resp_normalized.rename(columns={'time': 'time_resp', 'text': 'text_resp'}, inplace=True)
```

```python
# se concatena en nuevo dataframe
df_florida = pd.concat([df_florida, df_resp_normalized], axis=1)
df_florida
```

```python
# Se elimina la columna "reviews" del DataFrame resultante
df_florida = df_florida.drop(columns='resp')
df_florida
```

## Variable time

```python
def convertir_fecha(milisegundos_unix):       
    if milisegundos_unix == 0:
        return 0
    else:
        segundos = milisegundos_unix / 1000
        fecha = datetime.utcfromtimestamp(segundos)
        fecha_formateada = fecha.strftime('%Y-%m-%d %H:%M:%S')
        return fecha_formateada
```

```python
df_florida['time'] = df_florida['time'].apply(convertir_fecha)
df_florida
```

## Variable time_resp

```python
def convertir_a_entero(valor):
    return int(valor)
```

```python
df_florida['time_resp'] = df_florida['time_resp'].apply(convertir_a_entero)
df_florida
```

```python
df_florida['time_resp'] = df_florida['time_resp'].apply(convertir_fecha)
df_florida
```

```python
# Se Completa los valores que digan None en texto con el valor "SinDato"
df_florida['time_resp'] = df_florida['time_resp'].replace(0, 'SinDato')
df_florida
```
