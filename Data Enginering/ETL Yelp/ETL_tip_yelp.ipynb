```python
import pandas as pd
```

Importamos el archivo tip anterior

```python
df_tip_yelp = pd.read_csv('tip.csv')
df_tip_yelp.head(1)
```

```python
df_tip_yelp.shape
```

Importamos el archivo 'dataset_business_final.csv'

```python
df_business_yelp = pd.read_csv('dataset_business_final.csv')
df_business_yelp.head(1)
```

Nos quedamos con la columna unicamente de business_id de nuestros archivo business

```python
df_business_yelp = df_business_yelp[['business_id','ubicacion', 'name']]
df_business_yelp
```

Hacemos el merge para quedarnos unicamente con los tip que sean de negocios en Florida

```python
df_merged = pd.merge(df_business_yelp, df_tip_yelp, on='business_id', how='inner')
```

```python
df_merged.head(5)
```

```python
df_merged.shape
```

Nos quedamos con las columnas finales

```python
dataset_tip_final = df_merged[['Unnamed: 0','user_id','business_id','text',	'date',	'compliment_count']]
dataset_tip_final
```

Importamos el archivo ya limpio

```python
dataset_tip_final.to_csv('dataset_tip_final.csv', index=False)
```
