```python
import pandas as pd
```

Importamos el archivo user anterior

```python
df_user_yelp = pd.read_parquet('user.parquet')
df_user_yelp.head(1)
```

```python
df_user_yelp.shape
```

Importamos el datasete review_yelp

```python
df_review_yelp = pd.read_csv('dataset_review_final.csv')
df_review_yelp.head(1)
```

```python
df_review_yelp.shape
```

Importamos el datasete tip_yelp

```python
df_tip_yelp = pd.read_csv('dataset_tip_final.csv')
df_tip_yelp.head(1)
```

```python
df_tip_yelp.shape
```

Extraemos la columna 'user_id' de cada DataFrame

```python
user_id_tip = df_tip_yelp[['user_id']]
user_id_review = df_review_yelp[['user_id']]
```

```python
concatenar_user_id = pd.concat([user_id_review, user_id_tip])
concatenar_user_id
```

Borramos duplicados

```python
user_id= concatenar_user_id.drop_duplicates()
user_id
```

Hacemos elk mergue con el archivo anterior y este nuevo para filtrar por unicamente los usuarios con review en Florida

```python
df_merged = pd.merge(user_id, df_user_yelp, on='user_id', how='inner')
df_merged
```

Exporrtamos el archivo final

```python
dataset_review_final = df_merged
```

```python
dataset_review_final.to_csv('dataset_user_final.csv', index=False)
```
