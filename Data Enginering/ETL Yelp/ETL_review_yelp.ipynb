```python
import pandas as pd
```

Importamos el archivo review anterior

```python
df_review_yelp = pd.read_csv('dataset_review.csv')
df_review_yelp.head(1)
```

```python
df_review_yelp.shape
```

Importamos el archivo 'dataset_business_final.csv'

```python
df_business_yelp = pd.read_csv('dataset_business_final.csv')
df_business_yelp.head(1)
```

Nos quedamos con la columna unicamente de business_id de nuestros archivo business

```python
df_business_yelp = df_business_yelp[['business_id','ubicacion', 'name']]
df_business_yelp
```

Hacemos el merge para quedarnos unicamente con las review que sean de negocios en Florida

```python
df_merged = pd.merge(df_business_yelp, df_review_yelp, on='business_id', how='inner')
```

```python
df_merged.head(5)
```

```python
df_merged.shape
```

Nos quedamos con las columnas finales

```python
dataset_review_final = df_merged[['review_id', 'user_id','business_id',	'stars','useful','funny','cool','text',	'date']]
dataset_review_final
```

Importamos el archivo ya limpio

```python
dataset_review_final.to_csv('dataset_review_final.csv', index=False)
```
